```python
batsman=pd.read_csv('D:/Pandas/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Virat Kohli vs Rohut Sharma Career Comparison')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('V Kohloi vs R Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.title('V Kohloi vs R Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.title('V Kohloi vs R Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green')
plt.title('V Kohloi vs R Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot')
plt.title('V Kohloi vs R Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot',marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',marker='D')
plt.title('V Kohloi vs R Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot',marker='+')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',marker='+')
plt.title('V Kohloi vs R Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashdot',marker='D',markersize=10,label='V Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='dashdot',marker='D',markersize=10,label='RG Sharma')
plt.title('V Kohloi vs R Sharma Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dashdot',marker='D',markersize=10,label='V Kohli',color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashdot',marker='D',markersize=10,label='RG Sharma',color='green')
plt.title('V Kohli vs RG Sharma career comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='best')
```

```python
df=pd.read_csv('D:/Pandas/batsman_season_record.csv')
df
```

```python
np.arange(df.shape[0])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```
